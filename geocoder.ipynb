```python
import pandas as pd
from geopy.geocoders import Nominatim

nom=Nominatim()
df= pd.read_csv('supermarkets.csv')
df['Address']=df['Address']+", "+df['City']+", "+df['Country']
df
df['Coordinates']=df['Address'].apply(nom.geocode)
df['Coordinates']
df.Coordinates[0]
print("To get the latitude of the given address")
df['latitudes']= df['Coordinates'].apply(lambda x:x.latitude if x!=None else None)
df['longitudes']= df['Coordinates'].apply(lambda x:x.longitude if x!=None else None)
df

```
